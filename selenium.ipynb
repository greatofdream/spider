# 使用 selenium 避开反爬虫

表明原因在于https和http之间的差别，但是https用 `urllib` 并不能正常工作，因此还是需要selenium

```python
from selenium import webdriver
import urllib.request

keyword = '回族明星'
keywordCoding = urllib.request.quote(keyword)
baiduUrl = 'https://www.baidu.com/s?wd='
browser = webdriver.Chrome()
browser.get(baiduUrl + keywordCoding)
```

```python
elem = browser.find_elements_by_css_selector('p.op_exactqa_item_img')
```

```python
type(elem[1])
```

```python
for e in elem:
    e
```

```python
html = browser.page_source
```

```python
from bs4 import BeautifulSoup 
import requests
soup =  BeautifulSoup(html, 'html.parser')
img = soup.find_all(attrs='op_exactqa_item_img')
for im in img:
    print(im.img['src'],im.a['title'])
    r = requests.get(im.img['src'])
    print(r.content)
    with open(keyword+'/'+im.a['title']+'.png','wb') as opt:
        opt.write(r.content)
```
